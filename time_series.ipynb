```python
import numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import fetch_california_housing

california_housing = fetch_california_housing(as_frame=True)
df = california_housing.frame
```

```python
df.head(10)
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
reported_year = 1990
df['Date'] = pd.to_datetime((reported_year - df['HouseAge']), format='%Y')
```

```python
df.head()
```

```python
df.sort_values(by='Date', inplace=True)
df.head()
```

```python
df.index = pd.to_datetime(df['Date'], format='%Y-%m-%d')
del df['Date']
df.head()
```

```python
df.tail()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set()

plt.ylabel('Median house value')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.plot(df.index, df['MedHouseVal'], )
```

```python
data = df['MedHouseVal']
train_data, test_data = data[:int(len(data)*0.75)], data[int(len(data)*0.75):]
```

```python
plt.plot(train_data, color = "black")
plt.plot(test_data, color = "red")
plt.ylabel('Median house value')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.title("Train/Test split for California Housing dataset")
plt.show()
```

```python
train_data.shape
```

```python
test_data.shape
```

```python
from pmdarima.arima import auto_arima

model_autoARIMA = auto_arima(train_data, start_p=0, start_q=0,
                           test='adf',
                           max_p=7, max_q=7, m=1,
                           start_P=0, seasonal=False,
                           d=0, D=0, trace=True,
                           error_action='ignore',  
                           suppress_warnings=True, 
                           stepwise=True)

print(model_autoARIMA.aic())
```

```python
from statsmodels.tsa.arima.model import ARIMA

model = ARIMA(train_data, order = (6, 0, 2))
fitted = model.fit()
```

```python
test_data.shape

fc = fitted.forecast(len(test_data), alpha=0.05)
fc.index = test_data.index

plt.figure(figsize = (30, 5), dpi = 100)
plt.plot(train_data, label = 'training')
plt.plot(test_data, color = 'blue', label = 'Actual Median House Value')
plt.plot(fc, color = 'orange', label = 'Predicted Median House Value')

plt.title('Median House Value Prediction')
plt.xlabel('Time')
plt.ylabel('Actual Median House Value')
plt.legend(loc='upper left', fontsize=8)
plt.show()
```

```python
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(test_data, fc)
print('MSE: ' + str(mse))
mae = mean_absolute_error(test_data, fc)
print('MAE: ' + str(mae))
rmse = math.sqrt(mean_squared_error(test_data, fc))
print('RMSE: ' + str(rmse))
mape = np.mean(np.abs(fc - test_data)/np.abs(test_data))
print('MAPE: ' + str(mape))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose

result = seasonal_decompose(df['MedHouseVal'], model='multiplicative', period=30)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(16,9)
```

```python
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title('1st Order Differencing')
ax1.plot(df['MedHouseVal'].diff())

ax2 = f.add_subplot(122)
plot_acf(df['MedHouseVal'].diff().dropna(), ax=ax2)
plt.show()
```

```python
from statsmodels.tsa.stattools import adfuller

result = adfuller(df['MedHouseVal'].diff().dropna())
print('p-value: ', result[1])
```

```python
from matplotlib import pyplot as plt
from statsmodels.graphics.tsaplots import plot_acf

f = plt.figure()
ax1 = f.add_subplot(121)
ax1.set_title('2nd Order Differencing')
ax1.plot(df['MedHouseVal'].diff().diff())

ax2 = f.add_subplot(122)
plot_acf(df['MedHouseVal'].diff().diff().dropna(), ax=ax2)
plt.show()
```
